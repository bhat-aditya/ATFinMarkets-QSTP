```python
import numpy as np
import matplotlib.pyplot as plt

# IPython command for displaying plots in line
%matplotlib inline

x = np.arange(0, 10, 0.1)
y = np.sin(x)
plt.plot(x, y)
```

```python
# Specifying legends, plotting multiple lines in a single graph

y_sin = y
y_cos = np.cos(x)
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('X-axis values')
plt.ylabel('Y-axis values')
plt.title('Sin and Cos graphs')
plt.legend(['Sin', 'Cos'])
```

```python
# 3-D graphs

ax = plt.axes(projection = '3d')
ax.plot3D(np.sin(x), np.cos(x), x)
```

```python
# Subplot (Plotting different things in the same figure)

# Creating subplot grid having height 2 and width 1, setting 1st subplot as active
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title('Sin')

# Setting 2nd subplot as active
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cos')
plt.show()
```
